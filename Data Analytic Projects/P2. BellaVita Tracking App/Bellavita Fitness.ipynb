## Data Description
This dataset is from `Bellabeat Fitness Tracking App` about the user activity on their application. <br>
This the **`Capstone Project for the Google Data Analytics Professional Certificate program`** to provide the insights from the data.

## Column Description 
`Id`	:Unique identifier for each user. <br>
`ActivityDate`  :	The date of recorded activity (YYYY-MM-DD format).<br>
`TotalSteps`   :	The total number of steps taken on the given date.<br>
`TotalDistance` :	The total distance (in miles or km) covered on the given date.<br>
`TrackerDistance` :	Distance recorded by the fitness tracker (could be same as TotalDistance).<br>
`LoggedActivitiesDistance`:	Distance covered in logged activities (e.g., workouts, runs).<br>
`VeryActiveDistance` :	Distance covered during very active periods (e.g., running, intense workouts).<br>
`ModeratelyActiveDistance` :	Distance covered during moderate activity (e.g., brisk walking).<br>
`LightActiveDistance` :	Distance covered during light activity (e.g., casual walking).<br>
`SedentaryActiveDistance`:	Distance recorded while sedentary (usually minimal or zero).<br>
`VeryActiveMinutes` :	Total minutes spent in very active exercise.<br>
`FairlyActiveMinutes` :	Total minutes spent in fairly active exercise.<br>
`LightlyActiveMinutes` :	Total minutes spent in light activity.<br>
`SedentaryMinutes` :	Total minutes spent being sedentary (e.g., sitting or inactive).<br>
`Calories`:	The total number of calories burned on the given date.<br>

## Performing Data Wrangling 
Data wrangling is the process of converting and mapping data from its initial raw form to another format in order to make it more appropriate and valuable for a variety of  purposes such as analysis, visualization, and modeling.

```python
# import neccesary library 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read csv 
df = pd.read_csv("daily_activity.csv")
df.head()
```

```python
# Chacking out the null values in the column 
df.isnull().sum()
```

```python
# Checking for duplicated rows
df.duplicated().sum()
```

```python
# Changing the data type to Date time
df["ActivityDate"] = pd.to_datetime(df["ActivityDate"])
```

## Visualization   

```python
# DistinctCount of values 
df["Id"].nunique()
```

```python
# Find out the od with high step count 
df.groupby("Id")["TotalSteps"].sum().sort_values(ascending=False).head(8).plot(kind="barh")
plt.title("Top ID'd with HighSteps Count ")
plt.xlabel("Total Steps")
plt.ylabel("Id")
plt.show()
```

```python
# plotting histogram
plt.style.use("default")
plt.figure(figsize=(7,5)) # specify size of the chart
plt.hist(df["ActivityDate"].dt.day_name(), bins = 7, 
         width = 0.6, color = "blue", edgecolor = "black")

# adding annotations and visuals
plt.xlabel("Day of the week")
plt.ylabel("Frequency (Times)")
plt.title("Number of time user login in a week")
plt.show()
```

```python
# plotting scatter plot
plt.style.use("default")
plt.figure(figsize=(8,6)) # specify size of the chart
plt.scatter(df.TotalSteps, df.Calories, 
            alpha = 0.8, c = df.Calories, 
            cmap = "viridis")

# add annotations and visuals
median_calories = 2303
median_steps = 7637

plt.colorbar(orientation = "vertical")
plt.axvline(median_steps, color = "Blue", label = "Median steps")
plt.axhline(median_calories, color = "Red", label = "Median calories burned")
plt.xlabel("Steps taken")
plt.ylabel("Calories burned")
plt.title("Calories burned for every step taken")
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.style.use("default")
plt.figure(figsize=(8,6)) # Specify size of the chart
plt.scatter(df.TotalSteps, df.Calories, 
            alpha = 0.8, c = df.Calories, 
            cmap = "viridis")

# adding annotations and visuals
median_calories = 2303
median_hours = 20
median_sedentary = 991 / 60

plt.colorbar(orientation = "vertical")
plt.axvline(median_hours, color = "Blue", label = "Median steps")
plt.axvline(median_sedentary, color = "Purple", label = "Median sedentary")
plt.axhline(median_calories, color = "Red", label = "Median hours")
plt.xlabel("Hours logged")
plt.ylabel("Calories burned")
plt.title("Calories burned for every hour ")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# calculating total of individual minutes column
very_active_mins = df["VeryActiveMinutes"].sum()
fairly_active_mins = df["VeryActiveMinutes"].sum()
lightly_active_mins = df["LightlyActiveMinutes"].sum()
sedentary_mins = df["SedentaryMinutes"].sum()

# plotting pie chart
slices = [very_active_mins, fairly_active_mins, lightly_active_mins, sedentary_mins]
labels = ["Very active minutes", "Fairly active minutes", "Lightly active minutes", "Sedentary minutes"]
explode = [0, 0, 0.1, 0]
plt.style.use("default")
plt.pie(slices, labels = labels, 
         wedgeprops = {"edgecolor": "black"}, 
        explode = explode, autopct = "%1.1f%%")
plt.title("Percentage of Activity in Minutes")
plt.tight_layout()
plt.show()
```
